```python
import pandas as pd
from pycaret.classification import *
from pycaret.regression import *
from pycaret.datasets import get_data
```

```python
from pycaret.datasets import get_data

def load_data(dataset_name):

    data = get_data(dataset_name)
    return data

data = load_data('heart_disease')
```

```python
def train_ml_model(data, target_column, task='auto'):

    if task == 'auto':
        # Automatically determine whether to use regressors or classifiers based on the data type
        if data[target_column].dtype == 'object':
            task = 'classification'
        else:
            task = 'regression'

    # Setup PyCaret
    if task == 'classification':
        setup(data, target=target_column)
    elif task == 'regression':
        setup(data, target=target_column)

    # Compare models
    best_model = compare_models()

    return best_model

best_model = train_ml_model(data,'Disease')
print(best_model)
```

```python
def evaluate_ml_model(model, data):

    evaluate_model(model)

# Evaluate model
evaluate_ml_model(best_model, data)
```

```python
import streamlit as st


def main():
    st.title("Machine Learning Web App")

    # Allow users to upload a CSV file
    uploaded_file = st.file_uploader("Upload your file")
    if uploaded_file is not None:
        data = get_data(uploaded_file)

        # Display uploaded data
        st.write("Uploaded Data:")
        st.write(data.head())

        # Select target variable
        target_variable = st.selectbox("Select Target Variable", options=data.columns)

        # Select task type (classification or regression)
        task_type = st.selectbox("Select Task Type", options=["Classification", "Regression"])

        # Setup PyCaret
        if task_type == "Classification":
            setup(data, target=target_variable)
        elif task_type == "Regression":
            setup(data, target=target_variable, silent=True, verbose=False)

        # Compare models
        st.write("Comparing Models...")
        best_model = compare_models()
        st.write("Best Model:", best_model)

if __name__ == "__main__":
    main()
```
